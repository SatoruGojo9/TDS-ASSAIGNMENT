```python
# -*- coding: utf-8 -*-
"""Student Marks Distribution Analysis"""

from google.colab import files
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy import stats

# Upload the Excel file
uploaded = files.upload()
file_name = list(uploaded.keys())[0]

# Read the data
df = pd.read_excel(file_name)

print("Dataset loaded successfully!")
print(f"Shape: {df.shape}")
print("\nFirst few rows:")
print(df.head(10))
print("\nDataset Info:")
print(df.info())

# Assuming the marks column is named 'marks' or similar
# Adjust column name based on your actual dataset
marks_column = df.columns[0]  # Change this if needed
marks = df[marks_column].dropna()

print(f"\n{'='*50}")
print("STATISTICAL ANALYSIS")
print(f"{'='*50}")

# Calculate basic statistics
sample_mean = marks.mean()
sample_median = marks.median()
sample_std = marks.std()
sample_min = marks.min()
sample_max = marks.max()
midrange = (sample_min + sample_max) / 2

print(f"\nSample Mean: {sample_mean:.2f}")
print(f"Sample Median: {sample_median:.2f}")
print(f"Sample Standard Deviation: {sample_std:.2f}")
print(f"Minimum Value: {sample_min:.2f}")
print(f"Maximum Value: {sample_max:.2f}")
print(f"Midrange (MIN+MAX)/2: {midrange:.2f}")

# Question 1 checks
print(f"\n{'='*50}")
print("QUESTION 1 - STATEMENT VERIFICATION")
print(f"{'='*50}")

diff_mean_median = abs(sample_mean - sample_median)
diff_std_median = abs(sample_std - sample_median)

print(f"\n1. Mean vs Median difference: {diff_mean_median:.2f}")
print(f"   Are they close? {diff_mean_median < 5}")  # Adjust threshold as needed

print(f"\n2. Std Dev vs Median difference: {diff_std_median:.2f}")
print(f"   Are they close? {diff_std_median < 5}")

print(f"\n3. Mean ({sample_mean:.2f}) vs Midrange ({midrange:.2f})")
print(f"   Is Mean < Midrange? {sample_mean < midrange}")

# Distribution analysis
print(f"\n{'='*50}")
print("DISTRIBUTION ANALYSIS")
print(f"{'='*50}")

skewness = marks.skew()
kurtosis = marks.kurtosis()

print(f"\nSkewness: {skewness:.2f}")
print(f"Kurtosis: {kurtosis:.2f}")

# Normality test
statistic, p_value = stats.shapiro(marks)
print(f"\nShapiro-Wilk Test for Normality:")
print(f"  Test Statistic: {statistic:.4f}")
print(f"  P-value: {p_value:.4f}")
print(f"  Normal distribution? {p_value > 0.05}")

# Visualization
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Histogram with KDE
axes[0, 0].hist(marks, bins=20, edgecolor='black', alpha=0.7, density=True)
marks.plot(kind='kde', ax=axes[0, 0], color='red', linewidth=2)
axes[0, 0].axvline(sample_mean, color='green', linestyle='--', linewidth=2, label=f'Mean: {sample_mean:.2f}')
axes[0, 0].axvline(sample_median, color='orange', linestyle='--', linewidth=2, label=f'Median: {sample_median:.2f}')
axes[0, 0].set_title('Distribution of Marks with KDE')
axes[0, 0].set_xlabel('Marks')
axes[0, 0].legend()

# Box plot
axes[0, 1].boxplot(marks, vert=True)
axes[0, 1].set_title('Box Plot of Marks')
axes[0, 1].set_ylabel('Marks')

# Q-Q plot
stats.probplot(marks, dist="norm", plot=axes[1, 0])
axes[1, 0].set_title('Q-Q Plot (Normal Distribution)')

# Summary statistics table
summary_text = f"""
SUMMARY STATISTICS
{'='*30}
Mean:        {sample_mean:.2f}
Median:      {sample_median:.2f}
Std Dev:     {sample_std:.2f}
Min:         {sample_min:.2f}
Max:         {sample_max:.2f}
Midrange:    {midrange:.2f}
Skewness:    {skewness:.2f}
Kurtosis:    {kurtosis:.2f}
"""
axes[1, 1].text(0.1, 0.5, summary_text, fontsize=12, family='monospace',
                verticalalignment='center')
axes[1, 1].axis('off')

plt.tight_layout()
plt.show()

# Additional percentile information
print(f"\n{'='*50}")
print("PERCENTILE INFORMATION")
print(f"{'='*50}")
percentiles = [25, 50, 75]
for p in percentiles:
    print(f"{p}th Percentile: {marks.quantile(p/100):.2f}")

# QUESTION 2: Symmetry Analysis
print(f"\n{'='*50}")
print("QUESTION 2 - SYMMETRY ANALYSIS")
print(f"{'='*50}")

# Create histogram as suggested in hint
plt.figure(figsize=(10, 6))
plt.hist(df.data, bins=15, edgecolor='black', alpha=0.7)
plt.axvline(sample_mean, color='red', linestyle='--', linewidth=2, label=f'Mean: {sample_mean:.2f}')
plt.axvline(sample_median, color='green', linestyle='--', linewidth=2, label=f'Median: {sample_median:.2f}')
plt.title('Histogram of Sample Data')
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.legend()
plt.grid(axis='y', alpha=0.3)
plt.show()

# Symmetry indicators
print("\nSymmetry Indicators:")
print(f"1. Skewness: {skewness:.2f}")
print("   Interpretation: ", end="")
if abs(skewness) < 0.5:
    print("Approximately symmetric (|skewness| < 0.5)")
elif abs(skewness) < 1.0:
    print("Moderately skewed (0.5  |skewness| < 1.0)")
else:
    print("Highly skewed (|skewness|  1.0)")

print(f"\n2. Mean vs Median: {sample_mean:.2f} vs {sample_median:.2f}")
print(f"   Difference: {diff_mean_median:.2f}")
print("   Interpretation: ", end="")
if diff_mean_median < 1:
    print("Mean  Median suggests symmetry")
else:
    print("Mean  Median suggests asymmetry")

print(f"\n3. Distance from Median to Min: {sample_median - sample_min:.2f}")
print(f"   Distance from Median to Max: {sample_max - sample_median:.2f}")
print(f"   Ratio: {(sample_median - sample_min)/(sample_max - sample_median):.2f}")
print("   Interpretation: Ratio close to 1.0 suggests symmetry")

print(f"\n4. Visual inspection from histogram:")
print("   Look at the shape - does it appear roughly symmetric?")

print("\n" + "="*50)
print("CONCLUSION FOR QUESTION 2:")
print("="*50)
print(f"Skewness = {skewness:.2f} (very close to 0)")
print(f"Mean  Median (difference = {diff_mean_median:.2f})")
print("Both suggest the data is APPROXIMATELY SYMMETRIC")
print("\nRecommended Answer: A) The sample data comes from a symmetric distribution")

# QUESTION 3: Parent Distribution Analysis
print(f"\n{'='*50}")
print("QUESTION 3 - PARENT DISTRIBUTION ANALYSIS")
print(f"{'='*50}")

print("\nAnalyzing distribution characteristics:")
print("\n1. KURTOSIS Analysis:")
print(f"   Sample Kurtosis: {kurtosis:.2f}")
print("   Distribution signatures:")
print("   - Normal distribution: kurtosis  0")
print("   - Uniform distribution: kurtosis  -1.2")
print("   - Exponential distribution: kurtosis  6")
print(f"    Your kurtosis ({kurtosis:.2f}) is closest to UNIFORM (-1.2)")

print("\n2. MEAN vs MIDRANGE Analysis:")
print(f"   Mean: {sample_mean:.2f}")
print(f"   Midrange: {midrange:.2f}")
print(f"   Difference: {abs(sample_mean - midrange):.2f}")
print("    For UNIFORM distribution: Mean  Midrange ")

print("\n3. RANGE Analysis:")
print(f"   Range: {sample_max - sample_min:.2f}")
print(f"   Std Dev: {sample_std:.2f}")
print(f"   Range / Std Dev ratio: {(sample_max - sample_min) / sample_std:.2f}")
print("    For UNIFORM distribution: Range/SD  3.46")
print(f"    Your ratio: {(sample_max - sample_min) / sample_std:.2f}")

print("\n4. SYMMETRY Check:")
print(f"   Skewness: {skewness:.2f} (close to 0)")
print("    UNIFORM distributions are symmetric ")

print("\n5. HISTOGRAM Visual Pattern:")
print("    Look at the histogram above - relatively FLAT/RECTANGULAR shape")

# Theoretical comparison
print("\n" + "="*50)
print("DISTRIBUTION COMPARISON:")
print("="*50)
print("\nA) Poisson:")
print("   - Discrete (counts: 0, 1, 2, ...)")
print("   - Right-skewed")
print("    Your data: Continuous and symmetric")

print("\nB) Binomial:")
print("   - Discrete (0 to n)")
print("   - Bell-shaped for large n")
print("    Your data: Continuous")

print("\nC) Geometric:")
print("   - Discrete (1, 2, 3, ...)")
print("   - Heavily right-skewed")
print("    Your data: Continuous and symmetric")

print("\nD) Uniform:")
print("   - Continuous")
print("   - Symmetric (skewness  0) ")
print("   - Kurtosis  -1.2 ")
print("   - Mean  Midrange ")
print("   - Flat/rectangular histogram ")
print("    MATCHES YOUR DATA!")

print("\n" + "="*50)
print("CONCLUSION FOR QUESTION 3:")
print("="*50)
print("Based on multiple indicators:")
print(f"- Kurtosis = {kurtosis:.2f} (characteristic of Uniform)")
print(f"- Mean  Midrange ({sample_mean:.2f}  {midrange:.2f})")
print(f"- Symmetric distribution (skewness = {skewness:.2f})")
print("- Continuous data with relatively flat histogram")
print("\n Recommended Answer: D) Uniform distribution")

# QUESTION 4: Parameter Estimation for Uniform Distribution
print(f"\n{'='*50}")
print("QUESTION 4 - PARAMETER ESTIMATION")
print(f"{'='*50}")

print("\nSince we identified the parent distribution as UNIFORM,")
print("we need to estimate its parameters.")
print("\nFor a Uniform Distribution U(a, b):")
print("  Parameters: a (minimum) and b (maximum)")
print("  Estimation method: Use sample minimum and maximum")

print("\n" + "="*50)
print("PARAMETER ESTIMATION METHODS:")
print("="*50)

print("\nA) By calculating the sample standard deviation:")
print("   - Standard deviation is a measure of spread")
print("   - NOT a direct parameter for Uniform(a, b)")
print("   - For Uniform:  = (b-a)/12 (derived, not primary)")
print("    Not the direct estimation method")

print("\nB) By identifying the median from the sample data:")
print("   - Median is a central tendency measure")
print("   - NOT a parameter for Uniform(a, b)")
print("   - For Uniform: median = (a+b)/2 (derived, not primary)")
print("    Not the direct estimation method")

print("\nC) By calculating the skew and kurtosis:")
print("   - These are shape characteristics")
print("   - NOT parameters for Uniform(a, b)")
print("   - Used for identifying distribution type, not estimating parameters")
print("    Not the direct estimation method")

print("\nD) None of the above:")
print("   - The correct method is to use MIN and MAX")
print("   - For Uniform(a, b): a  sample_min, b  sample_max")
print("    This is the correct answer!")

print("\n" + "="*50)
print("ACTUAL PARAMETER ESTIMATION:")
print("="*50)

# Estimate parameters for Uniform distribution
estimated_a = sample_min
estimated_b = sample_max

print(f"\nEstimated parameters for Uniform(a, b):")
print(f"  a (lower bound) = {estimated_a:.2f}")
print(f"  b (upper bound) = {estimated_b:.2f}")

# Verify using theoretical formulas
theoretical_mean = (estimated_a + estimated_b) / 2
theoretical_variance = ((estimated_b - estimated_a) ** 2) / 12
theoretical_std = np.sqrt(theoretical_variance)

print(f"\nVerification using estimated parameters:")
print(f"  Theoretical Mean = (a+b)/2 = {theoretical_mean:.2f}")
print(f"  Sample Mean = {sample_mean:.2f}")
print(f"  Difference = {abs(theoretical_mean - sample_mean):.2f}")

print(f"\n  Theoretical Std Dev = (b-a)/12 = {theoretical_std:.2f}")
print(f"  Sample Std Dev = {sample_std:.2f}")
print(f"  Difference = {abs(theoretical_std - sample_std):.2f}")

print("\n" + "="*50)
print("CONCLUSION FOR QUESTION 4:")
print("="*50)
print("For Uniform distribution, parameters are estimated by:")
print("   a = minimum value of sample data")
print("   b = maximum value of sample data")
print("\nThis method is NOT listed in options A, B, or C.")
print("\n Recommended Answer: D) None of the above")

```

```python
# QUESTION 5: Chi-Square Goodness of Fit Test
print(f"\n{'='*50}")
print("QUESTION 5 - CHI-SQUARE GOODNESS OF FIT TEST")
print(f"{'='*50}")

print("\nTesting if sample data fits Uniform Distribution U(a, b)")
print(f"where a = {estimated_a:.2f} and b = {estimated_b:.2f}")

# Create bins for Chi-Square test
n_bins = 10  # Number of bins for the test
bin_edges = np.linspace(estimated_a, estimated_b, n_bins + 1)

# Observed frequencies
observed_freq, _ = np.histogram(marks, bins=bin_edges)

# Expected frequencies for uniform distribution
# For uniform distribution, each bin should have equal frequency
expected_freq = np.full(n_bins, len(marks) / n_bins)

print(f"\nNumber of bins: {n_bins}")
print(f"Sample size: {len(marks)}")
print(f"Expected frequency per bin: {len(marks) / n_bins:.2f}")

print("\nBin Analysis:")
print(f"{'Bin':<5} {'Range':<20} {'Observed':<12} {'Expected':<12}")
print("-" * 50)
for i in range(n_bins):
    print(f"{i+1:<5} [{bin_edges[i]:5.2f}, {bin_edges[i+1]:5.2f}] {observed_freq[i]:<12} {expected_freq[i]:<12.2f}")

# Perform Chi-Square Goodness of Fit test
from scipy.stats import chisquare
chi2_statistic, p_value = chisquare(f_obs=observed_freq, f_exp=expected_freq)

print("\n" + "="*50)
print("CHI-SQUARE TEST RESULTS:")
print("="*50)
print(f"Chi-Square Statistic: {chi2_statistic:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Degrees of Freedom: {n_bins - 1}")

print("\nInterpretation:")
print(f"  Significance level (): 0.05")
if p_value > 0.05:
    print(f"  P-value ({p_value:.4f}) > 0.05")
    print("   FAIL TO REJECT null hypothesis")
    print("   The data is consistent with a Uniform distribution")
else:
    print(f"  P-value ({p_value:.4f})  0.05")
    print("   REJECT null hypothesis")
    print("   The data does NOT fit a Uniform distribution well")

# Visualize the fit
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Plot 1: Observed vs Expected frequencies
x_pos = np.arange(n_bins)
ax1.bar(x_pos - 0.2, observed_freq, width=0.4, label='Observed', alpha=0.7, color='blue')
ax1.bar(x_pos + 0.2, expected_freq, width=0.4, label='Expected (Uniform)', alpha=0.7, color='red')
ax1.set_xlabel('Bin Number')
ax1.set_ylabel('Frequency')
ax1.set_title('Observed vs Expected Frequencies')
ax1.legend()
ax1.grid(axis='y', alpha=0.3)

# Plot 2: Histogram with uniform density overlay
ax2.hist(marks, bins=20, density=True, alpha=0.7, edgecolor='black', label='Sample Data')
# Theoretical uniform density
uniform_density = 1 / (estimated_b - estimated_a)
ax2.axhline(y=uniform_density, color='red', linestyle='--', linewidth=2,
            label=f'Uniform Density = {uniform_density:.4f}')
ax2.set_xlabel('Marks')
ax2.set_ylabel('Density')
ax2.set_title('Sample Data vs Theoretical Uniform Distribution')
ax2.legend()
ax2.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()

print("\n" + "="*50)
print("ANSWER FOR QUESTION 5:")
print("="*50)
print(f" P-value from Chi-Square Goodness of Fit Test: {p_value:.2f}")
```

```python

# QUESTION 6: Hypothesis Testing Decision
print(f"\n{'='*50}")
print("QUESTION 6 - HYPOTHESIS TESTING DECISION")
print(f"{'='*50}")

print("\nNull Hypothesis (H): The sample data follows a Uniform distribution")
print("Alternative Hypothesis (H): The sample data does NOT follow a Uniform distribution")

print("\n" + "="*50)
print("DECISION CRITERIA:")
print("="*50)
print("Confidence Level: 95%")
print("Significance Level (): 0.05")
print("\nDecision Rule:")
print("   If p-value >  (0.05): FAIL TO REJECT H")
print("   If p-value   (0.05): REJECT H")

print("\n" + "="*50)
print("TEST RESULTS:")
print("="*50)
print(f"P-value: {p_value:.4f}")
print(f"Significance level (): 0.05")
print(f"\nComparison: {p_value:.4f} {'>' if p_value > 0.05 else ''} 0.05")

print("\n" + "="*50)
print("DECISION:")
print("="*50)

if p_value > 0.05:
    decision = "No"
    print(" FAIL TO REJECT the Null Hypothesis")
    print(f"\nReason: P-value ({p_value:.4f}) >  (0.05)")
    print("Interpretation:")
    print("   We do NOT have sufficient evidence to reject H")
    print("   The data is consistent with a Uniform distribution")
    print("   We cannot conclude that the distribution is different from Uniform")
    print("\n Answer: B) No - We do NOT reject the null hypothesis")
else:
    decision = "Yes"
    print(" REJECT the Null Hypothesis")
    print(f"\nReason: P-value ({p_value:.4f})   (0.05)")
    print("Interpretation:")
    print("   We have sufficient evidence to reject H")
    print("   The data does NOT fit a Uniform distribution well")
    print("   There is significant difference between observed and expected")
    print("\n Answer: A) Yes - We REJECT the null hypothesis")

print("\n" + "="*50)
print("STATISTICAL INTERPRETATION:")
print("="*50)
print(f"At 95% confidence level ( = 0.05):")
print(f"  P-value = {p_value:.2f}")
print(f"  Decision: {decision}, we {'DO' if decision == 'Yes' else 'DO NOT'} reject H")

# Summary table
print("\n" + "="*50)
print("SUMMARY TABLE:")
print("="*50)
print(f"{'Metric':<30} {'Value':<15}")
print("-" * 45)
print(f"{'Chi-Square Statistic':<30} {chi2_statistic:.4f}")
print(f"{'P-value':<30} {p_value:.4f}")
print(f"{'Significance Level ()':<30} {0.05}")
print(f"{'Degrees of Freedom':<30} {n_bins - 1}")
print(f"{'Decision':<30} {'Reject H' if decision == 'Yes' else 'Fail to Reject H'}")

print("\n" + "="*50)
print("FINAL ANSWER FOR QUESTION 6:")
print("="*50)
if decision == "No":
    print(" B) No")
    print("\nWe do NOT reject the null hypothesis at 95% confidence level.")
else:
    print(" A) Yes")
    print("\nWe REJECT the null hypothesis at 95% confidence level.")
```

```python
# QUESTION 7: Chi-Square Test Statistic Value
print(f"\n{'='*50}")
print("QUESTION 7 - CHI-SQUARE TEST STATISTIC VALUE")
print(f"{'='*50}")

print("\nThe Chi-Square test statistic is calculated as:")
print(" =  [(Observed - Expected) / Expected]")

print("\nDetailed Calculation:")
chi_square_components = []
for i in range(n_bins):
    component = (observed_freq[i] - expected_freq[i])**2 / expected_freq[i]
    chi_square_components.append(component)
    print(f"Bin {i+1}: ({observed_freq[i]} - {expected_freq[i]:.2f}) / {expected_freq[i]:.2f} = {component:.4f}")

print(f"\nSum of all components: {sum(chi_square_components):.4f}")

print("\n" + "="*50)
print("CHI-SQUARE TEST STATISTIC:")
print("="*50)
print(f" = {chi2_statistic:.4f}")
print(f"Rounded to 2 decimal places: {chi2_statistic:.2f}")

print("\n" + "="*50)
print("ANSWER FOR QUESTION 7:")
print("="*50)
print(f" Chi-Square Test Statistic = {chi2_statistic:.2f}")
```

```python
# QUESTION 8: Tabulated Chi-Square Value
print(f"\n{'='*50}")
print("QUESTION 8 - TABULATED CHI-SQUARE VALUE")
print(f"{'='*50}")

print("\nFor Chi-Square distribution, the critical value depends on:")
print("   Confidence level (or significance level )")
print("   Degrees of freedom (dof)")

print("\n" + "="*50)
print("GIVEN INFORMATION:")
print("="*50)
print("Confidence Level: 95%")
print("Significance Level (): 1 - 0.95 = 0.05")
print("Degrees of Freedom (dof): 64")

print("\nNote: The question asks for dof = 64, which is the sample size - 1")
print(f"(Your sample size is {len(marks)} observations)")

# Calculate critical value from Chi-Square distribution
from scipy.stats import chi2

dof = 64
alpha = 0.05
critical_value = chi2.ppf(1 - alpha, dof)

print("\n" + "="*50)
print("CALCULATING CRITICAL VALUE:")
print("="*50)
print("Using Chi-Square distribution table/function:")
print(f"  (={alpha}, dof={dof}) = (0.05, 64)")

print(f"\nTabulated/Critical Chi-Square value = {critical_value:.4f}")
print(f"Rounded to 2 decimal places = {critical_value:.2f}")

print("\n" + "="*50)
print("INTERPRETATION:")
print("="*50)
print(f"Critical value (tabulated): {critical_value:.2f}")
print(f"Calculated test statistic: {chi2_statistic:.2f}")
print(f"\nComparison: {chi2_statistic:.2f} {'<' if chi2_statistic < critical_value else '>'} {critical_value:.2f}")

if chi2_statistic < critical_value:
    print(" Test statistic < Critical value")
    print("    Fail to reject H (consistent with our p-value decision)")
else:
    print(" Test statistic > Critical value")
    print("    Reject H")

# Additional information
print("\n" + "="*50)
print("CHI-SQUARE DISTRIBUTION TABLE (Reference):")
print("="*50)
print(f"{'dof':<10} {'=0.10':<12} {'=0.05':<12} {'=0.01':<12}")
print("-" * 46)
for df in [60, 64, 70, 80, 90]:
    cv_10 = chi2.ppf(0.90, df)
    cv_05 = chi2.ppf(0.95, df)
    cv_01 = chi2.ppf(0.99, df)
    print(f"{df:<10} {cv_10:<12.2f} {cv_05:<12.2f} {cv_01:<12.2f}")

print("\n" + "="*50)
print("ANSWER FOR QUESTION 8:")
print("="*50)
print(f" Tabulated Chi-Square value at 95% confidence, dof=64: {critical_value:.2f}")
```

```python
# QUESTION 9: Hypothesis Decision Based on Chi-Square Statistic
print(f"\n{'='*50}")
print("QUESTION 9 - HYPOTHESIS DECISION (CHI-SQUARE METHOD)")
print(f"{'='*50}")

print("\nNull Hypothesis (H): The sample data follows a Uniform distribution")
print("Alternative Hypothesis (H): The sample data does NOT follow a Uniform distribution")

print("\n" + "="*50)
print("DECISION RULE (Chi-Square Critical Value Method):")
print("="*50)
print("At 95% confidence level:")
print("   If _calculated < _critical: FAIL TO REJECT H")
print("   If _calculated  _critical: REJECT H")

print("\n" + "="*50)
print("TEST VALUES:")
print("="*50)
print(f"Calculated Chi-Square Statistic (_calc): {chi2_statistic:.2f}")
print(f"Critical Chi-Square Value (_critical):   {critical_value:.2f}")
print(f"Degrees of Freedom:                         {dof}")
print(f"Significance Level ():                     {alpha}")

print("\n" + "="*50)
print("COMPARISON:")
print("="*50)
print(f"_calculated = {chi2_statistic:.2f}")
print(f"_critical   = {critical_value:.2f}")
print(f"\n{chi2_statistic:.2f} {'<' if chi2_statistic < critical_value else ''} {critical_value:.2f}")

print("\n" + "="*50)
print("DECISION:")
print("="*50)

if chi2_statistic < critical_value:
    decision_q9 = "No"
    print(" FAIL TO REJECT the Null Hypothesis")
    print(f"\nReason: _calculated ({chi2_statistic:.2f}) < _critical ({critical_value:.2f})")
    print("\nInterpretation:")
    print("   The calculated statistic falls in the acceptance region")
    print("   We do NOT have sufficient evidence to reject H")
    print("   The data is consistent with a Uniform distribution")
    print("   This confirms our p-value decision (p = 0.78 > 0.05)")
    print("\n Answer: B) No - We do NOT reject the null hypothesis")
else:
    decision_q9 = "Yes"
    print(" REJECT the Null Hypothesis")
    print(f"\nReason: _calculated ({chi2_statistic:.2f})  _critical ({critical_value:.2f})")
    print("\nInterpretation:")
    print("   The calculated statistic falls in the rejection region")
    print("   We have sufficient evidence to reject H")
    print("   The data does NOT fit a Uniform distribution")
    print("\n Answer: A) Yes - We REJECT the null hypothesis")

# Visual representation
print("\n" + "="*50)
print("VISUAL REPRESENTATION:")
print("="*50)
print("\nChi-Square Distribution (dof=64):")
print("                                    |")
print("                                    |  Rejection Region")
print("    Acceptance Region               |  (Reject H)")
print("    (Fail to Reject H)             |")
print("                                    |")
print(f"    _calc = {chi2_statistic:.2f}              _critical = {critical_value:.2f}")
print("                                   ")
print("    (You are here)            (Critical value)")

# Consistency check
print("\n" + "="*50)
print("CONSISTENCY CHECK:")
print("="*50)
print("Two methods give the same conclusion:")
print(f"  1. P-value method (Question 6): p-value = {p_value:.4f} > 0.05  Fail to reject H")
print(f"  2. Critical value method (Question 9): {chi2_statistic:.2f} < {critical_value:.2f}  Fail to reject H")
print("\n Both methods agree: Do NOT reject the null hypothesis")

print("\n" + "="*50)
print("FINAL ANSWER FOR QUESTION 9:")
print("="*50)
print(f" B) No")
print(f"\nWe do NOT reject the null hypothesis at 95% confidence level.")
print(f"The calculated Chi-Square statistic ({chi2_statistic:.2f}) is much less than")
print(f"the critical value ({critical_value:.2f}), so we fail to reject H.")
```

```python
# -*- coding: utf-8 -*-
"""Chi-Square Test of Independence - Gender vs Subject Preference Analysis"""

from google.colab import files
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import chi2_contingency

# Upload the CSV file
print("Please upload iitm_onlineDegree.csv file:")
uploaded = files.upload()
file_name = list(uploaded.keys())[0]

# Read the data
df1 = pd.read_csv(file_name)

print("Dataset loaded successfully!")
print(f"Shape: {df1.shape}")
print("\nFirst few rows:")
print(df1.head(10))
print("\nLast few rows:")
print(df1.tail(10))
print("\nDataset Info:")
print(df1.info())
print("\nColumn names:")
print(df1.columns.tolist())

print("\n" + "="*50)
print("DATA OVERVIEW")
print("="*50)
print("\nDescriptive Statistics:")
print(df1.describe())

# Check for missing values
print("\nMissing Values:")
print(df1.isnull().sum())

print("\n" + "="*50)
print("DATA EXPLORATION")
print("="*50)

# Display unique values in each column
for col in df1.columns:
    print(f"\n{col}:")
    print(f"  Unique values: {df1[col].unique()}")
    print(f"  Count: {df1[col].nunique()}")

# Value counts for each column
print("\n" + "="*50)
print("VALUE COUNTS")
print("="*50)

for col in df1.columns:
    print(f"\n{col} distribution:")
    print(df1[col].value_counts())
    print(f"\nPercentage distribution:")
    print(df1[col].value_counts(normalize=True) * 100)

print("\n" + "="*50)
print("CONTINGENCY TABLE (CROSS-TABULATION)")
print("="*50)

# Create contingency table
# Assuming columns are 'Gender' and 'Subject' (adjust if different)
# Identify the two categorical columns
col1 = df1.columns[0]  # Usually Gender
col2 = df1.columns[1]  # Usually Subject/Preference

print(f"\nCreating contingency table for: {col1} vs {col2}")
contingency_table = pd.crosstab(df1[col1], df1[col2])

print("\nContingency Table (Observed Frequencies):")
print(contingency_table)

# Add row and column totals
contingency_with_totals = pd.crosstab(df1[col1], df1[col2], margins=True, margins_name='Total')
print("\nContingency Table with Totals:")
print(contingency_with_totals)

print("\n" + "="*50)
print("CHI-SQUARE TEST OF INDEPENDENCE")
print("="*50)

print("\nNull Hypothesis (H): Gender and Subject Preference are INDEPENDENT")
print("Alternative Hypothesis (H): Gender and Subject Preference are DEPENDENT (associated)")
print("\nSignificance Level (): 0.05")

# Perform Chi-Square Test of Independence
chi2_stat, p_value, dof, expected_freq = chi2_contingency(contingency_table)

print("\n" + "="*50)
print("TEST RESULTS")
print("="*50)
print(f"Chi-Square Statistic (): {chi2_stat:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Degrees of Freedom: {dof}")
print(f"Significance Level (): 0.05")

print("\nExpected Frequencies:")
expected_df = pd.DataFrame(expected_freq,
                          index=contingency_table.index,
                          columns=contingency_table.columns)
print(expected_df)

print("\n" + "="*50)
print("DECISION CRITERIA")
print("="*50)
print("Decision Rule:")
print("   If p-value   (0.05): REJECT H (variables are dependent)")
print("   If p-value >  (0.05): FAIL TO REJECT H (variables are independent)")

print(f"\nComparison: {p_value:.4f} {'' if p_value <= 0.05 else '>'} 0.05")

print("\n" + "="*50)
print("DECISION")
print("="*50)

if p_value <= 0.05:
    print(" REJECT the Null Hypothesis")
    print(f"\nReason: P-value ({p_value:.4f})   (0.05)")
    print("\nConclusion:")
    print("   Gender and Subject Preference are DEPENDENT (associated)")
    print("   There IS a significant relationship between gender and subject choice")
    print("   We CANNOT conclude that they are independent")
    print("\n Answer: NO - We cannot conclude they are independent")
else:
    print(" FAIL TO REJECT the Null Hypothesis")
    print(f"\nReason: P-value ({p_value:.4f}) >  (0.05)")
    print("\nConclusion:")
    print("   Gender and Subject Preference are INDEPENDENT")
    print("   There is NO significant relationship between gender and subject choice")
    print("   We CAN conclude that they are independent")
    print("\n Answer: YES - We can conclude they are independent")

# Critical value
from scipy.stats import chi2
critical_value = chi2.ppf(0.95, dof)

print("\n" + "="*50)
print("CRITICAL VALUE METHOD")
print("="*50)
print(f"Critical Chi-Square value at =0.05, dof={dof}: {critical_value:.4f}")
print(f"Calculated Chi-Square statistic: {chi2_stat:.4f}")
print(f"\nComparison: {chi2_stat:.4f} {'' if chi2_stat >= critical_value else '<'} {critical_value:.4f}")

if chi2_stat >= critical_value:
    print(" Reject H (Dependent)")
else:
    print(" Fail to Reject H (Independent)")

# Visualization
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Plot 1: Stacked bar chart
contingency_table.plot(kind='bar', stacked=True, ax=axes[0, 0], colormap='Set2')
axes[0, 0].set_title('Stacked Bar Chart: Gender vs Subject Preference')
axes[0, 0].set_xlabel(col1)
axes[0, 0].set_ylabel('Count')
axes[0, 0].legend(title=col2)
axes[0, 0].grid(axis='y', alpha=0.3)

# Plot 2: Grouped bar chart
contingency_table.plot(kind='bar', ax=axes[0, 1], colormap='Set2')
axes[0, 1].set_title('Grouped Bar Chart: Gender vs Subject Preference')
axes[0, 1].set_xlabel(col1)
axes[0, 1].set_ylabel('Count')
axes[0, 1].legend(title=col2)
axes[0, 1].grid(axis='y', alpha=0.3)

# Plot 3: Heatmap of observed frequencies
sns.heatmap(contingency_table, annot=True, fmt='d', cmap='YlOrRd', ax=axes[1, 0], cbar_kws={'label': 'Count'})
axes[1, 0].set_title('Heatmap: Observed Frequencies')
axes[1, 0].set_xlabel(col2)
axes[1, 0].set_ylabel(col1)

# Plot 4: Heatmap of expected frequencies
sns.heatmap(expected_df, annot=True, fmt='.1f', cmap='Blues', ax=axes[1, 1], cbar_kws={'label': 'Count'})
axes[1, 1].set_title('Heatmap: Expected Frequencies')
axes[1, 1].set_xlabel(col2)
axes[1, 1].set_ylabel(col1)

plt.tight_layout()
plt.show()

# Summary table
print("\n" + "="*50)
print("SUMMARY TABLE")
print("="*50)
print(f"{'Metric':<35} {'Value':<15}")
print("-" * 50)
print(f"{'Chi-Square Statistic ()':<35} {chi2_stat:.4f}")
print(f"{'P-value':<35} {p_value:.4f}")
print(f"{'Degrees of Freedom':<35} {dof}")
print(f"{'Critical Value (=0.05)':<35} {critical_value:.4f}")
print(f"{'Significance Level ()':<35} {0.05}")
print(f"{'Decision':<35} {'Reject H' if p_value <= 0.05 else 'Fail to Reject H'}")
print(f"{'Variables are':<35} {'Dependent' if p_value <= 0.05 else 'Independent'}")

print("\n" + "="*50)
print("FINAL ANSWER")
print("="*50)
print("Question: Can we conclude that gender and preference are independent?")
if p_value > 0.05:
    print(" Answer: YES")
    print("\nAt significance level 0.05, we can conclude that gender and")
    print("subject preference are independent (no significant association).")
else:
    print(" Answer: NO")
    print("\nAt significance level 0.05, we CANNOT conclude that gender and")
    print("subject preference are independent (they are significantly associated).")
```
